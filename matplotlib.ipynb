```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
print(x)
print("\n")
print(y)
print( "\n")
print(z)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y, 'g+')
plt.plot(x,z, 'r--')
plt.show()
```

```python
plt.plot(x,y,'ro')
```

```python
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x)')
plt.show()
```

```python
fig=plt.figure()
axis=fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
axis.bar(languages,people)
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
ax.scatter(x,y)
ax.plot(x,z)
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
ax.scatter(x,y,z)
plt.show()
```
